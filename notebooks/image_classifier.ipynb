```python
import requests

api_key = "AIzaSyBlLj8_SLXaPEiTomYCNI88W9sTUShEMF8"
cse_id = "01ba7ba547ef84d14"
# query = "grizzly bear"



import os
def search_image_urls(query, target_num_images=150):

  file_path = "path/to/your/file.txt"

  if os.path.exists(query+'.txt'):
      with open(query+'.txt', "r") as file:
          image_urls = file.readlines()
          image_urls = [line.strip() for line in image_urls]

          # Print the image_urls
          print(image_urls)
      print("returning from file")
      return image_urls

  num_results_per_page = 10
  num_images_downloaded = 0
  start_index = 1
  image_urls = []
  while num_images_downloaded < target_num_images:
    url = f"https://www.googleapis.com/customsearch/v1?key={api_key}&cx={cse_id}&q={query}&searchType=image&start={start_index}"

    response = requests.get(url)
    data = response.json()

    # Extract image URLs from the response
    image_urls += [item['link'] for item in data['items']]

    num_images_downloaded = len(image_urls)
    # Update the start index for the next page
    start_index += num_results_per_page

    # Break the loop if no more results are available
    if 'queries' not in data or 'nextPage' not in data['queries']:
        break
    with open(query+'.txt', 'w') as file:
      for item in image_urls:
          file.write(item + '\n')
  return image_urls


```

```python
person_types = 'young','adult', "old"
path = Path('persons')
if not path.exists():
    path.mkdir()
    for o in person_types:
        dest = (path/o)
        dest.mkdir(exist_ok=True)
        results = search_image_urls( f'{o} person')
        download_images(dest, urls=results)
```

```python
people = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(128))
dls = people.dataloaders(path)
```

```python
dls.valid.show_batch(max_n=3, nrows=1)
```

```python
people = people.new(item_tfms=Resize(128, ResizeMethod.Squish))
dls2 = people.dataloaders(path)
```

```python
dls2.valid.show_batch(max_n=5)
```

```python
people = people.new(item_tfms=RandomResizedCrop(128, min_scale=0.3))
dls = people.dataloaders(path)
dls.train.show_batch(max_n=15, nrows=5, unique=True)
```

```python
people = people.new(item_tfms=Resize(128), batch_tfms=aug_transforms(mult=1)  )
dls = people.dataloaders(path)
dls.train.show_batch(max_n=15, nrows=3, unique=True)
```

```python
people = people.new(
    item_tfms=RandomResizedCrop(224, min_scale=0.5),
    batch_tfms=aug_transforms())
dls = people.dataloaders(path)
```

```python
dls.train.show_batch()
```

```python
dls = people.dataloaders(path)
learn = vision_learner(dls, resnet18, metrics=error_rate)
```

```python
learn.fine_tune(4)
```

```python
interp =  ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(k=5)
```

```python
cleaner = ImageClassifierCleaner(learn)
cleaner
```
